-- Data Cleaning --

```python
# Converting Purchase_Amount from String to Numeric
# Remove Non-Numeric characters in data set
if 'Purchase_Amount' in EC.columns:
  EC['Purchase_Amount'] = EC['Purchase_Amount'].str.replace('[^0-9.]', '', regex=True)
  EC['Purchase_Amount'] = pd.to_numeric(EC['Purchase_Amount'], errors='coerce')
```

```python
# Converting Time_of_Purchase to Date_time
if 'Time_of_Purchase' in  EC.columns:
  EC['Time_of_Purchase'] = pd.to_datetime(EC['Time_of_Purchase'], errors='coerce')
```

```python
# Inspecting Missing Values
missing_values = EC.isnull().sum()
print(missing_values)
print(missing_values[missing_values>0])
```

```python
EC.dropna(subset=['Purchase_Amount', 'Time_of_Purchase'], inplace=True)
```

-- ML --


```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.metrics import accuracy_score
from sklearn.ensemble import RandomForestClassifier
```

```python
# Droping rows
EC_model = EC.dropna(subset=['Customer_Satisfaction'].copy())
```

```python
target = 'Customer_Satisfaction'
```

```python
drop_columns = ['Customer_ID', target]
features = EC_model.drop(columns=drop_columns, errors='ignore')
```

```python
numeric_features = features.select_dtypes(include=[np.number])
catergorical_features = features.select_dtypes(exclude=[np.number])
```

```python
print('Numeric features:', numeric_features)
print('Catergorical features:',catergorical_features)
```

```python
X = features.copy()
Y = EC_model[target].copy()
```

```python
# Split data
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

```python
# Build Preprocess Pipelines
numeric_transformer = Pipeline(steps=[
    ('scaler', StandardScaler())
])
categorical_transformer = Pipeline(steps=[
    ('onehot', OneHotEncoder(handle_unknown='ignore'))
])
```

```python
preprocessor = ColumnTransformer(
    transformers=[
        ('num', numeric_transformer, numeric_features.columns),
        ('cat', categorical_transformer, catergorical_features.columns)
    ])
```

```python
# Creating a pipeline that combine Preprocess and Classifer
clf = Pipeline(steps=[('preprocessor', preprocessor),
                      ('classifier', RandomForestClassifier())])
clf.fit(X_train, Y_train)
Y_pred = clf.predict(X_test)
```

```python
# Evaluate the model
accuracy = accuracy_score(Y_test, Y_pred)
print(f'Accuracy: {accuracy:.2f}')
```

```python
# Confusion matrix
cm = confusion_matrix(Y_test, Y_pred)
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
plt.title('Confusion Matrix')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.plot(Y_test.values, label='Original Data')
plt.plot(Y_pred, label='Model Prediction')
plt.xlabel('Data Point Index')
plt.ylabel('Customer Satisfaction')
plt.title('Model Prediction vs. Original Data')
plt.legend()
plt.grid(True)
plt.show()
```
